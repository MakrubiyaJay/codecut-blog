Building a maintainable data science project is essential for keeping your work clear, extensible, and reproducible over time. Without a strong structure, projects tend to become messy, leading to duplicated code, complicated debugging, and frustration when new collaborators join

Establishing an effective structure from the beginning addresses many of these issues. However, designing that structure from scratch can be challenging.

## Introducing the Data Science Template {#introducing-the-data-science-template}

To solve this, I developed the [data science template](https://github.com/khuyentran1401/data-science-template), a ready-to-use framework that embeds industry best practices. This template results from my years refining how data science projects should be organized to be clean, transparent, and collaborative.

![](https://codecut.ai/wp-content/uploads/2023/06/introduction2.gif)

I created this repo years ago. I recently updated the template to include uv, a lightning-fast Python package installer, and refreshed versions of core libraries to ensure compatibility with the latest tools.

> ğŸ“š For comprehensive project structure guidelines and organization patterns, check out [Production-Ready Data Science](https://codecut.ai/production-ready-data-science/).



The tools used in this template are:

- [Poetry](https://python-poetry.org/): A tool for dependency management and packaging in Python that creates isolated environments.
- [uv](https://github.com/astral-sh/uv): An extremely fast Python package installer and resolver, an alternative to pip and poetry's installer.
- [pip](https://pip.pypa.io/en/stable/): The standard package installer for Python, used to install and manage libraries.
- [Hydra](https://hydra.cc/): A flexible framework to manage configurations for complex applications, particularly useful for data science and machine learning projects.
- [Pre-commit plugins](https://pre-commit.com/): A framework for managing and maintaining multi-language pre-commit hooks to ensure code quality.
- [Pdoc](https://pdoc.dev/): A simple yet powerful tool for generating API documentation for Python projects.

## Install Dependencies {#install-dependencies}

Depending on which dependency manager you selected when setting up the project, follow the instructions below:

| Tool | Speed | Virtual Environment Support | Ease of Use | Best For |
|------|-------|----------------------------|-------------|----------|
| pip | Standard speed | No | Very easy | Simple projects, maximum compatibility |
| Poetry | Moderate | Yes | Easy | Managing isolated environments and packaging |
| uv | Extremely fast | Yes | Easy | Fast installation and modern dependency management |

### pip {#pip}

[pip](https://pip.pypa.io/en/stable/) is the standard Python package installer.

First, create and activate a virtual environment:

````python
python -m venv .venv
source .venv/bin/activate  # On macOS/Linux
.venv\Scripts\activate     # On Windows
````

To install all dependencies:

```bash
pip install -r requirements-dev.txt
```

To install only production dependencies:

```bash
pip install -r requirements.txt
```

To run a Python script:

```bash
python3 src/process.py
```

### Poetry {#poetry}

[Poetry](https://codecut.ai/poetry-a-better-way-to-manage-python-dependencies) is a dependency management tool that simplifies the creation of isolated environments and manages your packages efficiently.

To activate the virtual environment:

```bash
poetry shell
```

To install all project dependencies from `pyproject.toml`:

```bash
poetry install
```

To install only production dependencies:

```bash
poetry install --only main
```

To add a new library:

```bash
poetry add <library-name>
```

To run a Python script:

```bash
poetry run python src/process.py
```

### uv {#uv}

[uv](https://codecut.ai/why-uv-might-all-you-need/) is an ultra-fast Python package installer and resolver.

To install all dependencies from pyproject.toml:

```bash
uv sync --all-extras
```

To install only production dependencies:

```bash
uv sync
```

To add a new package:

```bash
uv add <library-name>
```

To run a Python script:

```bash
uv run src/process.py
```

### Configure Ruff Settings

To get the most out of Ruff, configure it through your `pyproject.toml` file. This controls how Ruff analyzes and formats your code during pre-commit checks:

```toml
[tool.ruff]
# Exclude a variety of commonly ignored directories.
exclude = [...]

# Same as Black.
line-length = 88

[tool.ruff.lint]
ignore = ["E501", "E402"]
select = ["B","C","E","F","W","B9", "I", "Q"]

[tool.ruff.format]
quote-style = "double"
indent-style = "space"
skip-magic-trailing-comma = false

[tool.ruff.lint.mccabe]
max-complexity = 10
```

Key sections explained:

- `[tool.ruff]`: Basic settings like excluded directories and line length compatibility with Black
- `[tool.ruff.lint]`: Controls which lint rules to ignore and select for code analysis
- `[tool.ruff.format]`: Formatting preferences for consistent code style across your project
- `[tool.ruff.lint.mccabe]`: Complexity checking to automatically flag overly complex functions

## Manage Configuration Files with Hydra {#manage-configuration-files-with-hydra}

It is essential to avoid hard-coding as it gives rise to various issues:

- **Maintainability**: If values are scattered throughout the codebase, updating them consistently becomes harder
- **Reusability**: Hardcoding values limits the reusability of code for different scenarios
- **Security concerns**: Hardcoding sensitive information like passwords or API keys directly into the code can be a security risk
- **Testing and debugging**: Hardcoded values can make testing and debugging more challenging

Configuration files solve these problems by storing parameters separately from the code, which enhances code maintainability and reproducibility.

[Hydra](https://hydra.cc/) stands out as an excellent configuration management tool because of its impressive set of features, including convenient parameter access and command-line configuration override.

Suppose the "main.yaml" file located in the "config" folder looks like this:

```yaml
# config/main.yaml
process:
  use_columns:
  - col1
  - col2
model:
  name: model1
data:
  raw: data/raw/sample.csv
  processed: data/processed/processed.csv
  final: data/final/final.csv
```

Within a Python script, you can effortlessly access a configuration file by applying a single decorator to your Python function. By using the dot notation (e.g., `config.data.raw`), you can conveniently access specific parameters from the configuration file:

````python
import hydra
from omegaconf import DictConfig

@hydra.main(config_path="../config", config_name="main", version_base="1.2")
def process_data(config: DictConfig):
    """Function to process the data"""

    print(f"Process data using {config.data.raw}")
    print(f"Columns used: {config.process.use_columns}")

if __name__ == "__main__":
    process_data()
````

To override configuration from the command line:

```bash
python src/processs_data.py data.raw=sample2.csv
```

Output:

```bash
Process data using sample2.csv
Columns used: ['col1', 'col2']
```

## Add API Documentation {#add-api-documentation}

As data scientists often collaborate with other team members on projects, it is important to create comprehensive documentation for these projects. However, manually creating documentation can be time-consuming.

This template leverages [pdoc](https://pdoc.dev/) to generate API documentation based on the docstrings of your Python files and objects.

To generate static documentation:

```bash
pdoc src -o docs
```

To start documentation server (available at http://localhost:8080):

```bash
pdoc src --http localhost:8080
```

Navigate to [http://localhost:8080](http://localhost:8080/) to access the documentation in your web browser.
